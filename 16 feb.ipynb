Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

->DML stands for Data Manipulation Language, which is a subset of SQL (Structured Query Language) used to manage and manipulate the data stored in a database. DML statements allow you to insert, update, and delete data within database tables.

INSERT: The INSERT statement is used to add new records (rows) to a table. You provide the values for each column in the new row. For example:

sql
Copy code
INSERT INTO Employees (EmployeeID, FirstName, LastName, DepartmentID)
VALUES (1, 'John', 'Doe', 101);
UPDATE: The UPDATE statement is used to modify existing records in a table. You specify the columns and values to be updated and use a WHERE clause to identify which rows to update. For example:

sql
Copy code
UPDATE Employees
SET DepartmentID = 102
WHERE LastName = 'Doe';
DELETE: The DELETE statement is used to remove records from a table. Like with the UPDATE statement, you use a WHERE clause to specify the rows to be deleted. For example:

sql
Copy code
DELETE FROM Employees
WHERE LastName = 'Doe';
These DML statements allow you to manipulate the data stored within the database, enabling you to add, modify, or remove information as needed.


Q4. What is DQL? Explain SELECT with an example.

DQL stands for Data Query Language, which is a subset of SQL (Structured Query Language) used to retrieve data from a database. The primary DQL statement is SELECT, which is used to query and retrieve specific data from one or more tables.

SELECT: The SELECT statement retrieves data from a table or tables based on specified criteria. It allows you to retrieve specific columns and filter rows using conditions.

Example:

sql
Copy code
SELECT FirstName, LastName
FROM Employees
WHERE DepartmentID = 102;
In this example, the SELECT statement retrieves the first and last names of employees from the "Employees" table where the DepartmentID is 102. The result will be a list of employees who belong to the specified department.
