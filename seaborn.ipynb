Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot for different events and regions
sns.set(style="darkgrid")  # Set the style for the plot

# Create a line plot using seaborn's lineplot function
plt.figure(figsize=(10, 6))  # Set the figure size
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Set labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of FMRI Data for Different Events and Regions")

# Show the plot
plt.legend(loc="upper right")  # Add legend
plt.tight_layout()
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create subplots for the two box plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Box plot for 'age' grouped by 'pclass'
sns.boxplot(data=titanic_data, x='pclass', y='age', ax=axes[0])
axes[0].set_title('Box Plot of Age by Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')

# Box plot for 'fare' grouped by 'pclass'
sns.boxplot(data=titanic_data, x='pclass', y='fare', ax=axes[1])
axes[1].set_title('Box Plot of Fare by Pclass')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for the 'cut' column
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True)

# Set plot title and labels
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for the 'species' column
sns.pairplot(data=iris_data, hue='species', palette='Set1')
plt.title('Pair Plot of Iris Dataset by Species')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix for the heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, annot=True, fmt='.0f', cmap='YlGnBu')  # Format as a floating-point number with no decimal places
plt.title('Passenger Counts Heatmap (Flights Dataset)')
plt.show()
```
