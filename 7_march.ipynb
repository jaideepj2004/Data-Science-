Q3. Measure the three measures of central tendency for the given height data:
[178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

```python
import statistics

height_data = [178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5]

# Calculate the mean
mean_height = statistics.mean(height_data)

# Calculate the median
median_height = statistics.median(height_data)

# Calculate the mode
try:
    mode_height = statistics.mode(height_data)
    print(f"Mode height: {mode_height}")
except statistics.StatisticsError:
    print("There is no unique mode in the dataset.")

print(f"Mean height: {mean_height}")
print(f"Median height: {median_height}")
```

```python
# Q4. Find the standard deviation for the given data:
# [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

import numpy as np

height_data = [178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5]

# Calculate the standard deviation
std_deviation = np.std(height_data)

print(f"Standard Deviation: {std_deviation}")
```

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

-> more is the value of range, standard deviation and variance more is the deivation of the data from the average value . For example for the consider the below two data set 

```python
import numpy as np

# Data Set 1
data_set1 = [10, 11, 9, 10, 11, 9, 10, 11, 9, 10]

# Data Set 2
data_set2 = [2, 18, 3, 20, 1, 22, 5, 17, 6, 19]

# Calculate and print statistics for Data Set 1
range1 = np.ptp(data_set1)  # Peak-to-Peak (Range)
variance1 = np.var(data_set1)  # Variance
std_deviation1 = np.std(data_set1)  # Standard Deviation

print("Statistics for Data Set 1:")
print(f"Range: {range1}")
print(f"Variance: {variance1}")
print(f"Standard Deviation: {std_deviation1}")
print()

# Calculate and print statistics for Data Set 2
range2 = np.ptp(data_set2)  # Peak-to-Peak (Range)
variance2 = np.var(data_set2)  # Variance
std_deviation2 = np.std(data_set2)  # Standard Deviation

print("Statistics for Data Set 2:")
print(f"Range: {range2}")
print(f"Variance: {variance2}")
print(f"Standard Deviation: {std_deviation2}")


```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data Set 1
data_set1 = [10, 11, 9, 10, 11, 9, 10, 11, 9, 10]

# Data Set 2
data_set2 = [2, 18, 3, 20, 1, 22, 5, 17, 6, 19]

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Smoothed histogram for Data Set 1
plt.subplot(1, 2, 1)
sns.histplot(data_set1, kde=True, color='blue', bins=5)
plt.title('Smoothed Histogram for Data Set 1')
plt.xlabel('Value')
plt.ylabel('Density')

# Smoothed histogram for Data Set 2
plt.subplot(1, 2, 2)
sns.histplot(data_set2, kde=True, color='green', bins=5)
plt.title('Smoothed Histogram for Data Set 2')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

here we can clearly see that the data 2 which has more range, standard deviation and variance has more skewness than the first data and hence it more spread as well 

Q6. What is a Venn diagram?

A Venn diagram is a visual representation that uses overlapping circles to show the relationships and commonalities between different sets or groups of data.

```python
# Q7. For the two given sets A = (2,3,4,5,6,7) & B = (0,2,6,8,10). Find:
# (i) A B
# (ii) A  B

# Define the sets A and B
A = {2, 3, 4, 5, 6, 7}
B = {0, 2, 6, 8, 10}

# Find the intersection (A  B)
intersection_result = A.intersection(B)

# Find the union (A  B)
union_result = A.union(B)

# Print the results
print("Intersection (A  B):", intersection_result)
print("Union (A  B):", union_result)
```

Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.



mean is the one which is most affected by the outliners wheareas median is not much affected by the outliners and mode may be affected completely by the outliner or it may not get affected at all , range is completely affected by the outliners and standard devation and variance are also affected by the outliners

```python
import numpy as np

# Create a dataset with outliers
data = [12, 15, 18, 21, 24, 200]  # An outlier (200) is added to the dataset

# Calculate mean, median, variance, and standard deviation
mean_data = np.mean(data)
median_data = np.median(data)
variance_data = np.var(data)
std_deviation_data = np.std(data)

# Print the results
print("Original Data:")
print("Mean:", mean_data)
print("Median:", median_data)
print("Variance:", variance_data)
print("Standard Deviation:", std_deviation_data)

# Remove the outlier and calculate the measures again
data_no_outlier = [x for x in data if x < 100]  # Removing the outlier

mean_no_outlier = np.mean(data_no_outlier)
median_no_outlier = np.median(data_no_outlier)
variance_no_outlier = np.var(data_no_outlier)
std_deviation_no_outlier = np.std(data_no_outlier)

# Print the results without the outlier
print("\nData without Outlier:")
print("Mean:", mean_no_outlier)
print("Median:", median_no_outlier)
print("Variance:", variance_no_outlier)
print("Standard Deviation:", std_deviation_no_outlier)
```
