```python
# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
# scatter plot for age and fare columns in the titanic dataset.

# Import necessary libraries
import seaborn as sns
import plotly.express as px

# Step 1: Load the Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")

# Step 2: Create a scatter plot for "age" vs "fare" using Plotly Express
fig = px.scatter(
    titanic,          # Specify the dataset
    x="age",         # X-axis: "age" column
    y="fare",        # Y-axis: "fare" column
    title="Age vs Fare Scatter Plot",  # Set the plot title
    labels={"age": "Age", "fare": "Fare"}  # Set axis labels
)

# Step 3: Show the plot
fig.show()
```

```python
#Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

# Import necessary libraries
import seaborn as sns
import plotly.express as px

# Step 1: Load the "tips" dataset from Seaborn
tips = sns.load_dataset("tips")

# Step 2: Create a box plot using Plotly Express
fig = px.box(
    tips,          # Specify the dataset
    x="day",       # X-axis: "day" column
    y="total_bill",# Y-axis: "total_bill" column
    title="Box Plot of Total Bill by Day",  # Set the plot title
    labels={"day": "Day", "total_bill": "Total Bill"}  # Set axis labels
)

# Step 3: Show the plot
fig.show()



```

```python
# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
# the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
# column with the color parameter.

# Import necessary libraries
import seaborn as sns
import plotly.express as px

# Use the "tips" dataset from Seaborn
tips = sns.load_dataset("tips")

# Create a histogram using Plotly Express
fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",  # Use the "day" column for color
    pattern_shape="smoker",  # Use the "smoker" column for pattern shape
    title="Histogram of Total Bill by Sex",
    labels={"sex": "Sex", "total_bill": "Total Bill"}
)

# Show the plot
fig.show()

```

```python
# Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
# the color parameter.
# Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
# dimensions parameter.


# Import necessary libraries
import seaborn as sns
import plotly.express as px

# Use the "iris" dataset from Seaborn
iris = sns.load_dataset("iris")

# Create a scatter matrix plot using Plotly Express
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset by Species"
)

# Show the plot
fig.show()

```

```python
#Q5. What is Distplot? Using Plotly express, plot a distplot.

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset from Seaborn (you can replace this with your own dataset)
data = sns.load_dataset("tips")

# Create a distplot using Seaborn
sns.distplot(data["total_bill"], kde=True, rug=True)

# Set labels and title
plt.xlabel("Total Bill Amount")
plt.ylabel("Density")
plt.title("Distribution of Total Bill Amount with KDE")

# Show the plot
plt.show()
```
